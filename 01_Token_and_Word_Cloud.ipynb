# Create a tokenizer

Write a function `tokenize` that takes a string and returns a list of tokens. 

```python
import re
def tokenize(doc):
  pattern = r'[A-Za-z0-9]+'
  
  return re.findall(pattern,doc)
```

# Create a function to count frequencies

Write a function `count_frequencies` that takes a list of tokens and returns a dictionary that counts the number of times each token appeared in the list. 

For example: 

`count_frequencies(['hello','hello','world'])` 

should return 

`{'hello':2,'world':1}`.

```python
def count_frequencies(tokens):
  Not_in_tokens = ['a', 'the', 'of','The','is','And','and','to','as']
  counts = {}
  for i in tokens:
    if i in Not_in_tokens:
      counts [i] = 0
    elif i not in counts:
        counts [i] = 1
    else:
        counts [i] +=1
  return counts
count_frequencies(['hello','hello','world'])
```

# Generate a word cloud

Generate a word cloud from the Wikipedia article you selected. This code shouldn't need to be modified, you just need to run it.

```python
# use your tokenizer to generate tokens
tokens = tokenize(doc)
```

```python
# count frequencies using the function you built
frequencies = count_frequencies(tokens)
```

```python
# use this code to generate the word cloud
import wordcloud
cloud = wordcloud.WordCloud()
cloud.generate_from_frequencies(frequencies)
import matplotlib.pyplot as plt
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.figure()
```
